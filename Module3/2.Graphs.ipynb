```python
import pandas as pd
import matplotlib.pyplot as plt

#create 2D array of table
data=[ ['E001','M',34,123,'Normal',350],
        ['E002','F',36,167,'Overweight',450],
        ['E003','M',33,124,'Normal',180],
        ['E004','M',34,136,'Obesity',200],
        ['E005','F',31,123,'Underweight',150],
        ['E006','M',29,190,'Normal',350],
        ['E007','F',45,120,'Overweight',50]]

#dataframe created with above
df = pd.DataFrame(data,columns=['EmpID','Gender','Age','Sales','BMI','Income'])

#creating histogram
df.hist()

#display histogram
plt.show()
```

```python
#Column Chart

#Dataframe of previous code is used all 3 age, income, sales
df.plot.bar()
plt.show()

#plot between 2 attributes
plt.bar(df['Age'],df['Sales'])
plt.xlabel('Age')
plt.ylabel('Sales')
plt.show()
```

```python
#Box plot chart

print("Box plot of dataset")
df.plot.box()
plt.show()

print("Box plot of Income")
plt.boxplot(df['Income'])
plt.title("Income", color='red')
plt.show()
```

```python
#Pie Chart

print("Pie Chart of Age")
plt.pie(df['Age'], labels = {"A", "B", "C","D", "E", "F", "G"}, autopct='% 1.1f %%', shadow = True)
plt.show()

print("Pie Chart of Income")
plt.pie(df['Income'], labels = {"A", "B", "C", "D", "E", "F","G"},autopct ='% 1.1f %%', shadow = True)
plt.show()

print("Pie Chart of Sales")
plt.pie(df['Sales'], labels = {"A", "B", "C", "D", "E", "F", "G"},autopct ='% 1.1f %%', shadow = True)
plt.show()
```

```python

# scatter plot between income and age plt.scatter(df['Income], df['Age])
plt.show()

# scatter plot between income and sales
print("Scatter plot of Income and Sales")
plt.scatter(df['Income'], df['Sales'])
plt.show()

# scatter plot between sales and age
print("Scatter plot of Sales and Age")
plt.scatter(df['Sales'], df['Age'])
plt.show()
```

Graph using matplotlib and numpy

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np

x = np.linspace(0, 10)
print("Plot using trigonometry Sin(x)")
plt.plot(x, np.sin(x), color='blue')
```

```python
print("Plot using trigonometry Sin(x) and Cos(x)")
plt.plot(x, np.sin(x), color='blue')
plt.plot(x, np.cos(x), color='g')
```

```python
print("Plot using trigonometry Tan(x)")
plt.plot(x, np.tan(x), color='blue')
```
